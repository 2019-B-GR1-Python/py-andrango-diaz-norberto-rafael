```python
class Persona:
    __id_persona = ""
    __id_ciudad = ""
    __nombre_persona = ""
    __edad_persona=0
    
    #constructor
    
    def __init__(self, id_persona, id_ciudad, nombre_persona, edad_persona=0):
        self.__id_persona = id_persona
        self.__id_ciudad = id_ciudad
        self.__nombre_persona = nombre_persona
        self.__edad_persona = edad_persona
    
    #getters
    
    def get_id_persona(self): 
        return self.__id_persona
    def get_id_ciudad(self): 
        return self.__id_ciudad
    def get_nombre_persona(self): 
        return self.__nombre_persona
    def get_poblacion(self): 
        return self.__edad_persona
      
    #setters 
    
    def set_id_nombre(self, id_nombre): 
        self.__id_nombre = id_nombre
    def set_id_ciudad(self, id_ciudad): 
        self.__id_ciudad = id_ciudad
    def set_nombre_persona(self, nombre_persona): 
        self.__nombre_persona = nombre_persona
    def set_edad_persona(self, edad_persona): 
        self.__edad_persona = edad_persona
        
    #toString
        
    def __str__(self):
        return f"ID de persona: {self.__id_persona}\nID de ciudad: {self.__id_ciudad}\nNombre persona: {self.__nombre_persona}\nEdad: {self.__edad_persona}\n"    

```

```python
class Ciudad:
    __id_ciudad = ""
    __nombre_ciudad = ""
    __poblacion = 0
    
    #constructor
    
    def __init__(self, id_ciudad, nombre_ciudad, poblacion=0):
        self.__id_ciudad = id_ciudad
        self.__nombre_ciudad = nombre_ciudad
        self.__poblacion = poblacion
    
    #getters
    
    def get_id_ciudad(self): 
        return self.__id_ciudad
    def get_nombre_ciudad(self): 
        return self.__nombre_ciudad
    def get_poblacion(self): 
        return self.__poblacion
      
    #setters 
    
    def set_id_ciudad(self, id_ciudad): 
        self.__id_ciudad = id_ciudad
    def set_nombre_ciudad(self, nombre_ciudad): 
        self.__nombre_ciudad = nombre_ciudad
    def set_poblacion(self, poblacion): 
        self.__poblacion = poblacion
        
    #toString
        
    def __str__(self):
        return f"ID de ciudad: {self.__id_ciudad}\nNombre ciudad: {self.__nombre_ciudad}\nPoblacion: {self.__poblacion}\n"
    
```
